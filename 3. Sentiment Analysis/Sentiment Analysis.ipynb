```python
data_prep = pd.read_sql_table('gdelt_data_prep', engine )
sourceUrl = data_prep.SOURCEURL
sourceUrl.head()
```

```python
# function to get article.title and article.text
from newspaper import Article

def getTitle(link):
    try:
        article = Article(link)
        article.download()
        article.parse()
        return article.title
    except:
        print ("This is an error message!")
        return 0;

def getArticle(link):
    try:
        article = Article(link)
        article.download()
        article.parse()
        return article.text
    except:
        print ("This is an error message!")
        return 0;
```

```python

first = 8000
end = 10000

data = data_prep[first:end]

for i in range(first, end):
    print('Data', i+1)
    
    title_data = getTitle(data.SOURCEURL[i])
    
    print(title_data)
    
    content_data = getArticle(data.SOURCEURL[i])
    
    data.loc[i,'title']=title_data
    data.loc[i,'content']=content_data
    
    data_input = data[i-first:i-first+1]
    data_input['index']=i

    data_input.to_sql('gdelt_data_text', con=engine, if_exists='append', index=False)
```

```python
corpus = []
for index, row in news_data.iterrows():
    cleaned = row['content'].strip()
    if (cleaned != ''):
        corpus.append({'text':cleaned,'sentiment':row['sentimen']})

corpus
```

```python
data_berita = pd.DataFrame.from_dict(corpus)
news_corpus = data_berita[:250]
news_corpus
```

```python
X = news_corpus.text
y = news_corpus.sentiment
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
documents = news_corpus.text
```

```python
documents
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)
```

```python
true_k = 5
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
```
