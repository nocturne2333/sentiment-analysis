## Getting Data from SQL

```python
import pymysql
import pandas as pd
from pandas import DataFrame
from sqlalchemy import create_engine

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'gdelt_id_data'

'''
conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')
'''

# Create Connection to database
engine = create_engine('mysql+pymysql://'+user+':'+password+'@'+host+':'+port+'/'+database)
#engine = create_engine('mysql+pymysql://root: @localhost:3306/gdelt_id_data')

def run(sql):
    df = pd.read_sql_query(sql, engine)
    return df
```

```python
# create object for each table
events = pd.read_sql_table('gdelt_events', engine )
gkg = pd.read_sql_table('gdelt_gkg', engine )
```

## Descriptive Statistic

```python
print("Shape of data events :", events.shape)
print("Shape of data gkg :", gkg.shape)
```

### - Shape of data
The **events** data consist of `283.792` rows and `58` columns

The **gkg** data consist of `70.580` rows and `15` columns

### - Total data recorded each month

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))

# get events months
months = events.DATEADDED.apply(str).str[0:6]

# plot data each months
events.groupby(months).size().plot(kind='bar')
```

```python
fig, ax = plt.subplots(figsize=(10,5))

# get gkg months
months = gkg.DATE.apply(str).str[0:6]

# plot data each months
gkg.groupby(months).size().plot(kind='bar')
```

### - Total data recorded each day

```python
fig, ax = plt.subplots(figsize=(17,10))

# plot data each months
events.groupby('DATEADDED').size().plot(kind='bar')
```

```python
fig, ax = plt.subplots(figsize=(17,10))

# plot data each months
gkg.groupby('DATE').size().plot(kind='bar')
```

## Clean Data
In order to work with the data more easily, the tone column from the datasets will be called as sentiment. The sentiment values range between -100 and +100, with 0 indicating neutral. This is calculated as Positive Score minus Negative Score. Before that, all the null value will be removed from the datasets

```python
sentiment = events[['AvgTone', 'Actor1CountryCode', 'Actor1Name']].copy()
sentiment.groupby('Actor1Name').mean().sort_values('AvgTone')
```
